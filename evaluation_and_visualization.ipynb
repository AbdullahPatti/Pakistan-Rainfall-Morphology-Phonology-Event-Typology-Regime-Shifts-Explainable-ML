```python
# evaluation_and_visualization.ipynb

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve, auc

# ────────────────────────────────────────────────
#   Load results from modeling (you must have these)
# ────────────────────────────────────────────────

# Example – replace with your actual variables
results_dict = {
    'Logistic Regression': {'f1': 0.72, 'auc': 0.9715},
    'Random Forest':       {'f1': 0.93, 'auc': 0.9994},
    'ANN':                 {'f1': 0.91, 'auc': 0.9992},
}

# Replace with your real prediction arrays
predictions_dict = {
    'Logistic Regression': np.random.randint(0,7,913),   # ← YOUR real predictions
    'Random Forest':       np.random.randint(0,7,913),
    'ANN':                 np.random.randint(0,7,913),
}

# You need real y_test too for confusion matrix
y_test = np.random.randint(0,7,913)   # ← YOUR real test labels

# Regime shift files (from previous notebook)
shift_df = pd.read_csv('detected_regime_shifts.csv')
yearly_props_df = pd.read_csv('yearly_typology_proportions_long.csv')


# ────────────────────────────────────────────────
#   Plotting functions (same as you had)
# ────────────────────────────────────────────────

def plot_model_comparison(results_dict):
    models = list(results_dict.keys())
    f1_scores = [results_dict[m]['f1'] for m in models]
    auc_scores = [results_dict[m]['auc'] for m in models]

    x = np.arange(len(models))
    width = 0.35

    fig, ax = plt.subplots(figsize=(10, 6))
    ax.bar(x - width/2, f1_scores, width, label='Macro F1')
    ax.bar(x + width/2, auc_scores, width, label='ROC-AUC')

    ax.set_xlabel('Models')
    ax.set_ylabel('Score')
    ax.set_title('Model Performance Comparison')
    ax.set_xticks(x)
    ax.set_xticklabels(models, rotation=45)
    ax.legend()

    plt.tight_layout()
    plt.show()


def plot_confusion_matrices(y_test, predictions_dict):
    n_models = len(predictions_dict)
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))
    axes = axes.ravel()

    for idx, (model_name, y_pred) in enumerate(predictions_dict.items()):
        if idx < 4:
            cm = confusion_matrix(y_test, y_pred)
            sns.heatmap(cm, annot=True, fmt='d', ax=axes[idx], cmap='Blues')
            axes[idx].set_title(model_name)
            axes[idx].set_xlabel('Predicted')
            axes[idx].set_ylabel('Actual')

    plt.tight_layout()
    plt.show()


def plot_regime_shifts(shift_df, yearly_props_df):
    fig, axes = plt.subplots(3, 3, figsize=(15, 12))
    axes = axes.ravel()

    districts_to_plot = shift_df['district'].head(9).tolist()
    
    for idx, district in enumerate(districts_to_plot):
        district_data = yearly_props_df[yearly_props_df['adm2_id'] == district]
        shift_year = shift_df[shift_df['district'] == district]['shift_year'].values[0]

        ax = axes[idx]
        for typ in district_data['rainfall_type'].unique():
            type_data = district_data[district_data['rainfall_type'] == typ]
            label = f"Type {typ}"
            ax.plot(type_data['year'], type_data['proportion'],
                   label=label, marker='o', markersize=3)

        ax.axvline(x=shift_year, color='red', linestyle='--', linewidth=2, label='Regime Shift')
        ax.set_title(f'District {district}')
        ax.set_xlabel('Year')
        ax.set_ylabel('Proportion')
        ax.legend(loc='upper right', fontsize='small')

    plt.tight_layout()
    plt.show()


# ────────────────────────────────────────────────
#   Generate the plots
# ────────────────────────────────────────────────

print("Generating model comparison plot...")
plot_model_comparison(results_dict)

print("\nGenerating confusion matrices...")
plot_confusion_matrices(y_test, predictions_dict)

print("\nGenerating regime shift plots...")
plot_regime_shifts(shift_df, yearly_props_df)

print("\nVisualization complete.")
```
